### Collecting data from API and using requests module

```python
import requests
API_key='21d60ebd8917558baa95b132'
source_currency = input("Please enter the source currency ex INR: ")
amount = float( input("Please enter the amount to be converted: "))
target_currency = input("Please enter the target currency ex USD: ")

base_url = f'https://v6.exchangerate-api.com/v6/{API_key}/latest/{source_currency}'
```

```python
response = requests.get(base_url)
response
```

### Showcasing data in proper dictionary format which Json provide

```python
data = response.json()
data
```

```python
conversion_rate = data['conversion_rates']
conversion_rate
```

### Performing conversion rate operations

```python
if target_currency in conversion_rate:
    exchange_rate= conversion_rate[target_currency]
    converted_amount = amount*exchange_rate
    print(f" the converted amount is {converted_amount} of {target_currency}") 
else:
    print("The currency not found")
```
